```python
node_dict = {}
for line in inp:
    try:
        before, after = line.split(' -> ')
    except:
        before = line.split(' -> ')[0]
        after = ''
    
    parent, weight = re.match(r'([a-zA-Z]+) \(([0-9]+)\)', before).groups()
    children = [w.strip() for w in after.split(',')]
    
    node_dict[(parent, int(weight))] = children

# maps (node parent word, weight): [child1,..]
node_dict
```

```python
class Node:
    def __init__(self, name, weight, children):
        self.name = name
        self.weight = weight
        self.children = children
        self.sub_weights = []

    def __str__(self):
        return '%s, (%s)- [%s], [%s]' % (self.name, self.weight, ','.join(self.children), ', '.join(str(x) for x in self.sub_weights))
# maps word: Node
node_class_dict = {}
for (parent, weight), children in node_dict.items():
    n = Node(parent, weight, children)
    node_class_dict[parent] = n

print(node_class_dict.keys())
```

```python
def find_total_weight(nodestr):
    if not nodestr:
        return 0
    node = node_class_dict[nodestr]
    weight = 0
    for n in node.children:
        try:
            weight += find_total_weight(n)
        except KeyError:
            continue
    return weight + node.weight

start = node_class_dict[top_word]
find_total_weight(top_word)
```

```python
def record_subweights(startstr):
    if not startstr:
        return
    node = node_class_dict[startstr]
    for child in node.children:
        node.sub_weights.append(find_total_weight(child))
        record_subweights(child)

def print_nodes(startstr):
    if not startstr:
        return
    node = node_class_dict[startstr]
    print(node)
    for child in node.children:
        print_nodes(child)
        
record_subweights(top_word)
print_nodes(top_word)
```
